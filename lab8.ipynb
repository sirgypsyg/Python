```python
import matplotlib.pyplot as plt
import numpy as np
import time

c = -0.4 + 0.6j
#c = -0.5251993 + 0.5251993j
#JULIAa
x = []
y = []
colors = []

def julia_set(z, c):
    for i in range(300):
        z = z**2 + c
        if abs(z) > 2:
            return i
    return 0

d = 400
start_t = time.time()
for i in range(d):
    for j in range(d):
        xi = -1.5+3/d*i
        yi = -1.5+3/d*j
        x.append(xi)
        y.append(yi)
        colors.append(julia_set(xi+yi*1j, c))
print(time.time() - start_t)
plt.figure(figsize=(15,15))
plt.scatter(x, y, c = colors, s=0.1, cmap='prism')
#plt.scatter(x, y, c = colors, s=0.4, cmap='prism')
plt.axis('square')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import time

c = -0.1 + 0.65
d = 400

X, Y = np.meshgrid(
    np.linspace(-1.5, 1.5, d),
    np.linspace(-1.5, 1.5, d)
)

def julia_set(z, c):
    for i in range(300):
        z = z**2 + c
        colors.append(i)  # Append the calculated values of "i" to the "colors" array
    return z

start_t = time.time()
colors = []  # Initialize the "colors" array
julia_set(X+Y*1j, c)  # Call the julia_set function to populate the "colors" array
print(time.time() - start_t)

plt.figure(figsize=(10,10))
plt.scatter(X, Y, c=colors, s=0.4)  # Use X and Y directly instead of empty arrays "x" and "y"
plt.axis('square')
plt.show()
```

```python
c=-0.5251993+-0.5251993j

def julia_set(z, c):
    for i in range(300):
        z = z**2 + c
        if abs(z) > 2:
            return i
    return 300

X, Y = np.meshgrid(
    np.linspace(-1.5, 1.5, d),
    np.linspace(-1.5, 1.5, d)
)

Z = X + Y*1j
start_t = time.time()

julia = np.frompyfunc(julia_set, 2, 1)(Z, c).astype(np.float64)
print(time.time() - start_t)
plt.figure(figsize=(10,10))
plt.contourf(X, Y, julia)
plt.axis('square')
plt.show()
```

```python
from matplotlib.animation import FuncAnimation
import numpy as np
fig, ax = plt.subplots()
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)

points = ax.plot([], [], "ro")

point = points[0]

def animate(frame):
    x = frame
    y = 0
    points.set_data([x], [y])
    return [point]

animation = FuncAnimation(
    fig,
    animate,
    frames = np.arange(-10, 10, 0.1)
)

plt.show()
```
