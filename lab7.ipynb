```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(0, 1, 51)
#plt.figure(figsize=(12,10))
for a in np.linspace(0.25, 2, 8):
    plt.plot(X, X**a, color=(a/2, 0, 1-a/2),
             label=f'$x^{str(a)}$') 

plt.legend()
plt.show()
```

```python
X, Y = np.meshgrid(np.linspace(-10, 10, 30), 
                   np.linspace(-20, 20, 30))
Z = X**2 + Y**2

plt.contourf(X, Y, Z, levels=40)
plt.show()
```

```python
c = -0.4 + 0.6j
#JULIAa
x = []
y = []
colors = []

def julia_set(z, c):
    for i in range(300):
        z = z**2 + c
        if abs(z) > 2:
            return i
    return 0

d = 1000
for i in range(d):
    for j in range(d):
        xi = -1.5+3/d*i
        yi = -1.5+3/d*j
        x.append(xi)
        y.append(yi)
        colors.append(julia_set(xi+yi*1j, c))

plt.figure(figsize=(15,15))
plt.scatter(x, y, c = colors, s=0.1, cmap='prism')
plt.axis('square')
plt.show()

```

```python

x = []
y = []
colors = []

def julia_set(p):
    z = 0 + 0.j
    for i in range(300):
        z = z**2 + p
        if abs(z) > 2:
            return i
    return 300

d = 1000
for i in range(d):
    for j in range(d):
        xi = -2.5+3.5/d*i
        yi = -1.5+3/d*j
        x.append(xi)
        y.append(yi)
        colors.append(julia_set(xi+yi*1j))

plt.figure(figsize=(15,15))
plt.scatter(x, y, c = colors, s=0.1, cmap='prism')
plt.axis('square')
plt.show()

```
